```python
# load kaggle GDP dataset and clean missing values
import pandas as pd
import numpy as np
import os
#print(os.getcwd())
# load dataset
df = pd.read_csv(r'c:\Users\cgx\My_projects\Power_BI\Lipstick_Effect_on_Books\data\raw\world_gdp_raw.csv')

# See dataset info
print(df.info())
print(df.describe())

# check for missing values
print(df.isnull().sum())
# check for duplicates
print(f"Duplicates: {df.duplicated().sum()}")

# Delete rows where years has values < 1000
df = df[df['Year'] >= 1000]
```

```python
# check column names
print(df.columns)

# drop unnecessary columns
df = df.drop(columns=['Entity', 'Code'])
# rename columns
df = df.rename(columns={'Year': 'year', 'World GDP in 2011 Int.$ (OWID based on World Bank & Maddison (2017))': 'gdp'})
# print column names
print(df.columns)
```

```python
# output cleaned dataset
df.to_csv(r'c:\Users\cgx\My_projects\Power_BI\Lipstick_Effect_on_Books\data\Processed\gdp_clean.csv', index=False)
```
