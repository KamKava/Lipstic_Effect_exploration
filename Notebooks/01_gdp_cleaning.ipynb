```python
# ----------------------------
#  1. Load and inspectthe data
# ----------------------------

# import libraries
import pandas as pd
import numpy as np

# load dataset
df = pd.read_csv('../Data/Raw/gdp_raw.csv')

# inspect the dataset
print(df.head())
print(df.info())
print(df.describe())
print(df.columns)
```

```python
# ============================
# 2. Drop unnecessary columns
# ============================

# filter rows where na_item is B16Q
df[df["na_item"] == "B16Q"]

# keep only columns ["geo", "TIME_PERIOD", "OBS_VALUE", "unit", "OBS_FLAG"]
df = df[["geo", "TIME_PERIOD", "OBS_VALUE", "unit"]]

# rename columns
df = df.rename(columns={
    "geo": "country",
    "TIME_PERIOD": "year",
    "OBS_VALUE": "gdp_per_capita",
    "unit": "unit"
})

# inspect the dataset again
print(df.head())
```

```python
# ======================
# 3. Convert data types
# ======================

# convert year to integer
df["year"] = df["year"].astype(int)

# convert gdp_per_capita to numeric, coercing errors to NaN
df["gdp_per_capita"] = pd.to_numeric(df["gdp_per_capita"], errors="coerce")

# check data types
print(df.dtypes)
```

```python
# ==========================
# 4. Handle missing values
# ==========================

# check for missing values
df = df.dropna(subset=["gdp_per_capita"])
print(df.isnull().sum())

# check for duplicates
duplicates = df.duplicated(subset=["country", "year"])
print("Duplicate rows:", duplicates.sum())

# sort by country and year
df = df.sort_values(by=["country", "year"])

# drop duplicates, keeping the first occurrence
df = df.drop_duplicates(subset=["country", "year"], keep="first")

print(df.head())
```

```python
# ========================
# 5. save and export the data
# ========================

# save cleaned dataset
df.to_csv('../Data/Processed/gdp_clean.csv', index=False)
```
